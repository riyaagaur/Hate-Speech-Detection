```python
# Hate Speech Detection - Exploratory Data Analysis (EDA)

# Step 1: Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Step 2: Read the dataset
train_df = pd.read_csv("train_E6oV3lV.csv")

# Step 3: Display basic dataset info
train_df.info()

# Step 4: Check class distribution (Hate Speech vs. Non-Hate Speech)
plt.figure(figsize=(6,4))
sns.countplot(x=train_df['label'], palette=['blue', 'red'])
plt.xticks([0, 1], ['Non-Hate Speech', 'Hate Speech'])
plt.xlabel("Tweet Category")
plt.ylabel("Count of Tweets")
plt.title("Class Distribution of Hate Speech Dataset")
plt.show()

# Step 5: Analyze tweet length distribution
train_df["tweet_length"] = train_df["tweet"].apply(lambda x: len(str(x).split()))
plt.figure(figsize=(6,4))
sns.histplot(train_df["tweet_length"], bins=30, kde=True, color='purple')
plt.xlabel("Tweet Length (Number of Words)")
plt.ylabel("Count of Tweets")
plt.title("Distribution of Tweet Lengths")
plt.show()

# Step 6: Generate Word Cloud for Non-Hate Speech (label=0)
non_hate_tweets = " ".join(train_df[train_df["label"] == 0]["tweet"])
wordcloud_non_hate = WordCloud(width=800, height=400, background_color="white").generate(non_hate_tweets)

plt.figure(figsize=(8,4))
plt.imshow(wordcloud_non_hate, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud - Non-Hate Speech")
plt.show()

# Step 7: Generate Word Cloud for Hate Speech (label=1)
hate_tweets = " ".join(train_df[train_df["label"] == 1]["tweet"])
wordcloud_hate = WordCloud(width=800, height=400, background_color="black", colormap="Reds").generate(hate_tweets)

plt.figure(figsize=(8,4))
plt.imshow(wordcloud_hate, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud - Hate Speech")
plt.show()

# Step 8: Display dataset summary statistics
train_df.describe()
```
